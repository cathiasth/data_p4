## 2. Le nombre théorique de personnes qui pourraient être nourries 

```python
dispoAliment = pd.read_csv('dispo_alimentaire.csv')
```

```python
dispoAliment.head()
```

```python
population = pd.read_csv('population.csv')
```

```python
population.head()
```

```python
population2017 = population[(population['Année'] == 2017)]
```

```python
population2017.head()
```

- Disponibilité alimentaire mondiale

```python
nbTheo = dispoAliment.merge(population2017, how='inner')
```

```python
nbTheo.head()
```

```python
nbTheo['Disponibilité alimentaire (kcal/personne/année)'] = nbTheo['Disponibilité alimentaire (Kcal/personne/jour)']*365
```

```python
nbTheo['Population']= nbTheo['Valeur']
```

```python
nbTheo = pd.DataFrame(nbTheo, columns=['Zone','Année','Population','Disponibilité alimentaire (Kcal/personne/jour)', 'Disponibilité alimentaire (kcal/personne/année)'])
```

```python
normekcal = 2500*365
```

```python
nbTheo['Portion annuelle'] =(nbTheo['Disponibilité alimentaire (kcal/personne/année)']*nbTheo['Population'])/normekcal
```

```python
nbTheo.head()
```

```python
nbTheo.groupby(['Zone', 'Année', 'Population']).sum()
```

```python
dispoVege['Population'] = dispoVege['Valeur']
```

```python
dispoVege = pd.DataFrame(dispoVege, columns=['Zone','Année','Population','Origine','Disponibilité alimentaire (Kcal/personne/jour)', 'Disponibilité alimentaire (kcal/personne/année)'])
```

```python
conditionVege = dispoVege[(dispoVege.Origine == 'vegetale') & (dispoVege.Année == 2017)]
```

```python
dispoVege = conditionVege
```

```python
dispoVege['Portion annuelle'] =(dispoVege['Disponibilité alimentaire (kcal/personne/année)']*dispoVege['Population'])/normekcal
```

```python
dispoVege.groupby(['Zone', 'Année','Population','Origine']).sum()
```
