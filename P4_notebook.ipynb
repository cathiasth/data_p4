```python
nbTheo['Portion annuelle'] =nbTheo['Disponibilité alimentaire (kcal/personne/année)']*nbTheo['Population']
nbTheo
```

```python
totalPortionAnnuelle = nbTheo['Portion annuelle'].sum()
totalPortionAnnuelle
```
